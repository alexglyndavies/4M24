```python
import numpy as np
import matplotlib.pyplot as plt
```

Vector space $V$ is the open unit circle. We formally define $V$ as.
$$
V = \{v\in \mathbb{R}^2: \Vert v\Vert_2 < 1\} 
$$




We define a sequence of points in this space $\{v_1, v_2, \dots, v_n, \dots \}$, where $v_n = \begin{bmatrix} \sqrt{\frac{1}{2} \left(1- \frac{1}{n}\right)} \\  \sqrt{\frac{1}{2} \left(1- \frac{1}{n}\right)} \end{bmatrix}$. This sequence begins at the origin, and tends towards the boundary. In the limit the point lies on the boundary, which is not in $V$, and therefore $V$ is incomplete.

$v = \underset{n\rightarrow \infty}{\lim}v_n =  \begin{bmatrix} \sqrt{\frac{1}{2}} \\  \sqrt{\frac{1}{2}} \end{bmatrix}$,  $\left\Vert v\right\Vert = 1 \notin V$

To prove $V$ is incomplete, we must show that this sequence is Cauchy, and if it is, then we have found a Cauchy sequence with elements in $V$, with a limit not in $V$ and hence we have proved $V$ is not complete.

$$
\Vert v_n - v_m \Vert = \left|\frac{1}{m} - \frac{1}{n}\right| \rightarrow 0 , \quad \text{when} \quad n, m \rightarrow \infty
$$ 
Therefore the sequence is a Cauchy sequence, and $V$ is incomplete.

```python
r = 1 # Unit circle
c = 1/np.sqrt(2)

# Defining the sequence v_n
v = lambda n: (c*np.sqrt(r - r/n), c*np.sqrt(r - r/n))
```

```python
figure, ax = plt.subplots(nrows=1,ncols=2, figsize=(20,10))
plt.tight_layout()



theta = np.linspace(0, 2*np.pi, 100)
radius = 1
a = radius*np.cos(theta)
b = radius*np.sin(theta)

draw_circle = plt.Circle((0, 0), 1, color='k', alpha=0.35)
ax[0].plot(a, b, linewidth=5, linestyle='--', color='k')
ax[0].add_artist(draw_circle)

points = 1000
vs = np.zeros((points,2))
for n in np.arange(points):
    p = v(n+1)
    vs[n][0] = p[0]
    vs[n][1] = p[1]

ax[0].scatter(vs[:,0],vs[:,1],color='k', s=500, marker='+')
    
ax[0].scatter(c,c,color='r',zorder=10, s=500, marker='+')

ax[0].axis('off')
ax[0].set_aspect(1)
ax[0].set_xlim((-1.1,1.1))
ax[0].set_ylim((-1.1,1.1))


draw_circle = plt.Circle((0, 0), 1, color='k', alpha=0.35)
ax[1].plot(a, b, linewidth=5, linestyle='--', color='k')
ax[1].add_artist(draw_circle)

ax[1].scatter(vs[:,0],vs[:,1],color='k', s=500, marker='+')

ax[1].scatter(c,c,color='r',zorder=10, s=500, marker='+')
# ax.axis('off')
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_aspect(1)
ax[1].set_xlim((c-0.1,c+0.1))
ax[1].set_ylim((c-0.1,c+0.1))
ax[1].annotate(r'$v_n =  \left[\sqrt{\frac{1}{2} \left(1- \frac{1}{n}\right)} ,  \sqrt{\frac{1}{2} \left(1- \frac{1}{n}\right)} \right]^{T}$',(c-0.04,c+0.05),fontsize=25)
ax[1].annotate(r'$= \underset{n\rightarrow \infty}{\lim} v_n \notin V$',(c+0.04,c),fontsize=25)
ax[1].annotate(r'$v$',(c+0.035,c),fontsize=25, color='r')
plt.show()
```
