```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1 align="center"><b><u>Lecture 14 - Parallel Tempering MCMC</u></b></h1>

This notebook demonstrates the Parallel Tempering MCMC algorithm, and its ability to generate samples from multimodal distributions. A simple example of a mixture of Gaussian distributions is used.

```python
sigma = 0.2
mu = 1.5
```

```python
def gaussian(x,mu,sigma):
    return 1/np.sqrt(2*np.pi*np.square(sigma))*np.exp(-(1/(2*np.square(sigma)))*np.square(x-mu))
```

We define the target distribution as $$ \pi(x) = \frac{1}{2}\mathcal{N}(x;\mu,\sigma^2) + \frac{1}{2}\mathcal{N}(x;-\mu,\sigma^2) $$

```python
def pi(x):
    global mu, sigma
    return (1/2)*(gaussian(x,mu,sigma)+gaussian(x,-mu,sigma))
```

The smoothed density $\pi_{\beta}(x) = \pi(x)^{\beta}$ is defined below.

```python
def pi_beta(x,beta):
    global pi
    return np.power(pi(x),beta)
```

We can visulise the different densities with varying levels of smoothing to see how this changes the denisty being targeted.

```python
lim = 5
n = 1000
betas = [1,0.5,0.2,0.05]

x = np.linspace(-lim,lim,n)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(15,5))

# Function for plotting color based on temperature of distribution
c = lambda t: (1-1/(t+0.01),0.0,1/(t+0.01))

for i in range(len(betas)):
    y = [pi_beta(xi,betas[i]) for xi in x]
    ax.plot(x,np.array(y)/(2*sum(y)*lim/n),label=r"$\beta = {}$".format(betas[i]),color=c(1/betas[i]),lw=2.5)
plt.xlim((-4,4))
plt.ylim((0,1.1))
plt.title('Tempering Multimodal Distribution',fontsize=20)
plt.tight_layout()
plt.legend()
plt.show()
```

The target denisty has a large region of low probability around zero, and so if the step-size of a Metropolis-Hastings algorithm is smaller than this region, the sampler will almost never transition between the two modes of the distribution. For the higher temperature distributions (lower $\beta$) the gap is smoothed over.

To illustrate this, a MH sampler is implemented with a proposal $q(x,y)$ defined as a uniform distribution with a step-size $\gamma$ as follows:

$$
q(x|y) = \mathcal{U}\left(x;y-\frac{\gamma}{2},y+\frac{\gamma}{2}\right)
$$

```python
def acceptance_prob(x_new, x_prev, target_density):
    return min(1, target_density(x_new)/target_density(x_prev))

# Proposal defined as a uniform distribution centered on previous sample, width of stepsize
def proposal(x, stepsize):
    return np.random.uniform(low=x - 0.5 * stepsize, 
                             high=x + 0.5 * stepsize)

def step_MH(x_prev, target_density, stepsize):
    global proposal
    
    # Propose new sample
    x_new = proposal(x_prev, stepsize)
    
    # Accept or reject
    accept = np.random.rand() < acceptance_prob(x_new, x_prev, target_density)
    if accept:
        return x_new, accept
    else:
        return x_prev, accept
    
def MH(init, stepsize, n_total, target_density):
    
    acc = 0 
    samples = [init]
    
    for i in range(n_total):
        x_new, accept = step_MH(samples[-1], target_density, stepsize)
        samples.append(x_new)
        acc += accept
    
    return samples, acc/n_total
```

```python
init = 1.0
stepsize = 0.3
n_total = 20000

samples, acc = MH(init, stepsize, n_total, pi)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=len(betas),figsize=(20,5))
for i in range(len(betas)):
    target = lambda x: pi_beta(x, betas[i])
    
    samples, acc = MH(init, stepsize, n_total, target)
    
    ax[i].hist(samples,density='probability',bins=50)
    
    y = [target(xi) for xi in x]
    ax[i].plot(x,np.array(y)/(2*sum(y)*lim/n),color=c(1/betas[i]),lw=2.5)
    ax[i].set_title(r"$\beta = {}$".format(betas[i]),fontsize=20)
```

The figure above shows the MH samplers running on different temperature distributions independently.

Clearly for low temperature the sampler fails to explore both the modes of the distribution, buth with higher temperature, the full smoothed distributions are explored.

We now define the Parallel Tempering algorithm, which runs Markov chains on different temperature distributions, where attempted swapping between adjacent temperature chains is used as a proposal mechanism, with the aim of the smoother distributions informing samples from the target and allowing sampling from targets with multiple modes. 

The scheme for choosing the chains to swap we use is the 'deterministic even-odd' scheme. On an 'even' iteration, the pairs will begin at 0, and will attempt to swap with the adjacent higher temperature.

```python
# Example to show how the deterministic even-odd scheme attempts swaps
indices = np.arange(8)
evens = indices[::2]
odds = indices[1::2]

swaps_even = list(zip(evens, odds))
swaps_odd = list(zip(evens[1:], odds))

print("Even Swap Pairs: ",swaps_even)
print("Odd Swap Pairs: ",swaps_odd)
```

This scheme is used in practice as it allows for efficient parallel computation of acceptance probabilities, as the acceptance probabilities between these swap pairs are independent, and the switching between odd and even pairs allows for swaps to diffuse between all temperatures.

```python
### Thanks to Andrius Ovsianas for this implementation

# Code for generating list of target distributions
def make_target(beta):
    return lambda x : pi_beta(x, beta)

def PT(init, betas, stepsize, n_total):
    
    n_chains = len(betas)
    
    # Generate lists of even and odd indicies
    indices = np.arange(n_chains)
    evens = indices[::2]
    odds = indices[1::2]

    # Define the target distributions \pi_{\beta}
    targets = [make_target(beta) for beta in betas]

    # Create empty array for samples and initialise first samples at init
    samples = np.empty((n_total + 1, n_chains), dtype=np.float)
    samples[0] = init

    # Parallel Tempering algorithm with deterministic even-odd swap scheme
    for i in range(n_total):

        # Iterate through the each chain
        for j in range(n_chains):
            # MH sample from each chain independently
            samples[i + 1, j], _ = step_MH(samples[i, j], targets[j], stepsize)

        # Define swap pairs based on even or odd iteration
        if i % 2 == 0:
            swaps = list(zip(evens, odds))
        else:
            swaps = list(zip(evens[1:], odds))

        # For each pair of swaps, accept or reject a swap
        for a, b in swaps:
            alpha = targets[a](samples[i + 1][b]) * targets[b](samples[i + 1][a]) / (targets[a](samples[i + 1][a]) * targets[b](samples[i + 1][b]))
            alpha = min(1, alpha)

            accept = np.random.rand() < alpha
            if accept:
                # If accepted, swap the samples
                samples[i + 1, a], samples[i + 1, b] = samples[i + 1, b], samples[i + 1, a]
                
    return samples
            
```

```python
samples = PT(init, betas, stepsize, n_total)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=len(betas),figsize=(20,5))
for i in range(len(betas)):
    target = lambda x : pi_beta(x, betas[i])
    ax[i].hist(samples[:, i],density='probability',bins=50)
    
    y = [target(xi) for xi in x]
    ax[i].plot(x,np.array(y)/(2*sum(y)*lim/n),color=c(1/betas[i]),lw=2.5)
    ax[i].set_title(r"$\beta = {}$".format(betas[i]),fontsize=20)
    ax[i].set_xlim((-lim,lim))
plt.show()
```

The PT sampler is able to produce samples from both modes of the target distribution, which is a result of the swaps diffusing from higher temperature distributions and proposing samples from both modes. In practice, step-sizes are tuned to give optimal sampling based on MH acceptance rates, and this would improve the sampling of the higher temperature distribution (here the step-size is too low for $\beta=0.05$ and this results in the two modes of $\pi_{0.05}$ not being evenly explored as there is poor mixing).
